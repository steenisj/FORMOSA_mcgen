```python
import numpy as np
import matplotlib.pyplot as plt
import os
import math
```

```python
def yVsEta(eta, m, pt):
    temp = np.sqrt(m*m+pt*pt*np.cosh(eta)*np.cosh(eta))
    temp = temp + pt*np.sinh(eta)
    temp = temp/np.sqrt(m*m+pt*pt)
    return np.log(temp)
    

def dYdEta(eta, m, pt, h=0.01):
    diff = yVsEta(eta+h, m, pt)-yVsEta(eta-h, m, pt)
    return diff/(2*h)
```

```python
mass = [3.096916, 9.46030]
ptrans = np.linspace(0,30,31)
fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111)
ax.plot(ptrans, dYdEta(0.11, mass[0], ptrans), label=r'$J/\psi$')
ax.plot(ptrans, dYdEta(0.11, mass[1], ptrans), label=r'$\Upsilon$(1S)', color='red')
ax.set_xlim(0,30)
ax.set_ylim(0,1.2)
ax.grid()
ax.set_xlabel('$P_T$ (Gev)', fontsize='xx-large')
ax.set_ylabel(r'$\frac{d \eta}{d y}$', fontsize='xx-large', rotation=0, labelpad=15)
ax.legend(fontsize='xx-large')
plt.savefig('deta-dy.pdf')
```
